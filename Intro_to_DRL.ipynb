```python
import gym
env = gym.make("FrozenLake-v1", render_mode='rgb_array', is_slippery=False)
env.reset()
plt.figure()
for t in range(10):
   print("\nTimestep {}".format(t))
   
   plt.imshow(env.render())
   plt.show()

   a = env.action_space.sample()
   action_space = {0:'Left', 1:'Down', 2:'Right', 3:'Up'}

   # # Take the action and get the new observation space
   new_obs, reward, done, trunc, info = env.step(a)

  #  trunc, ob, r, done, info = env.step(a)
   print('-------------------------')
   print('Move:',action_space[a])
   print('Done:', done)

   if done:
      plt.imshow(env.render())
      plt.show()
      print("\nEpisode terminated early")
      break
      env.close()
```
